# Poisson Equation
## The Differential Equation
$$  \frac{\partial^2 u}{\partial y^2} + \frac{\partial^2 u}{\partial x^2}=f(x,y)$$

$$  \frac{\partial^2 u}{\partial y^2} + \frac{\partial^2 u}{\partial x^2}=0$$


## Boundary Condition
$$ u(x,0)=\sin(2\pi x), $$
$$ u(x,1)=\sin(2\pi x), $$
$$ u(0,y)=2\sin(2\pi y), $$
$$ u(1,y)=2\sin(2\pi y), $$

## The Difference Equation
$$ w[i-1,j]+ w[i,j-1] -4w[i,j]+w[i,j+1]+w[i+1,j]=0$$


```python
# LIBRARY
# vector manipulation
import numpy as np
# math functions
import math 

# THIS IS FOR PLOTTING

%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import warnings
warnings.filterwarnings("ignore")
```

```python
N=20
h=1/(N-1)


x=np.arange(0,1.0001,h)
y=np.arange(0,1.0001,h)

N2=(N-2)*(N-2)
A=np.zeros((N2,N2))
C=np.zeros((N,N))


f=np.zeros(N2)

w=np.zeros(N2)

b=np.zeros(N2)


for i in range (0,N):
        C[i,0]=np.sin(2*np.pi*x[i]) #left Boundary
        C[i,N-1]=np.sin(2*np.pi*x[i]) #Right Boundary

for j in range (0,N):
        C[0,j]=2*np.sin(2*np.pi*y[j]) #Lower Boundary
        C[N-1,j]=2*np.sin(2*np.pi*y[j]) #Upper Boundary
        
        
for i in range (0,N-2):
        b[i]=np.sin(2*np.pi*x[i+1]) #left Boundary
        b[i+(N-2)*(N-3)]=np.sin(2*np.pi*x[i+1])# Right Boundary
      
print(-b)
    
for j in range (0,N-2):
        b[(N-2)*j]=b[(N-2)*j]+2*np.sin(2*np.pi*y[j+1]) # Lower Boundary
        b[N-3+(N-2)*j]=b[N-3+(N-2)*j]+2*np.sin(2*np.pi*y[j+1])# Upper Boundary

                          
print(-b)

            
for i in range (0,N-2):
    for j in range (0,N-2):           
        A[i+(N-2)*j,i+(N-2)*j]=-4
        f[i+(N-2)*j]=h*h*0#(x[i+1]**2+y[j+1]**2)


        
for i in range (1,N-2):
    for j in range (0,N-2):           
        A[i+(N-2)*j,i+(N-2)*j-1]=1

for i in range (0,N-3):
    for j in range (0,N-2):           
        A[i+(N-2)*j,i+(N-2)*j+1]=1
      
for i in range (0,N-2):
    for j in range (0,N-3):           
        A[i+(N-2)*j,i+(N-2)*(j+1)]=1
      

for i in range (0,N-2):
    for j in range (1,N-2):           
        A[i+(N-2)*j,i+(N-2)*(j-1)]=1


fig = plt.figure(figsize=(8,4))
plt.matshow(A)
plt.show()
Ainv=np.linalg.inv(A)

w=np.dot(Ainv,f-b)
#print(x)
#print(y)
print(f)
print(b)
#print(f-b)


C[1:N-1,1:N-1]=w.reshape((N-2,N-2))
fig = plt.figure()
plt.imshow(C)
plt.xticks(np.arange(len(x)), x)
plt.yticks(np.arange(len(y)), y)
plt.xlabel('x')
plt.ylabel('y')
clb=plt.colorbar()
clb.set_label('w')

plt.show()

fig = plt.figure(figsize=(8,4))
plt.plot(w)
plt.xlabel('x')
plt.ylabel('w')
```
